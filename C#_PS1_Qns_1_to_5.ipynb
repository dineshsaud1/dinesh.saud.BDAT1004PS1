## Queston 3 C#
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is s(s - a)(s -b)(s -c), where s = (a+b+c)/2.

```python
internal class Program
{
    private static void Main(string[] args)
    {
        Console.Write("Enter the length of the first side of the triangle (a): ");
        double A = double.Parse(Console.ReadLine());

        Console.Write("Enter the length of the second side of the triangle (b): ");
        double B = double.Parse(Console.ReadLine());

        Console.Write("Enter the length of the third side of the triangle (c): ");
        double C = double.Parse(Console.ReadLine());

        double CalculateTriangleArea(double a, double b, double c)
        {
            double s = (a + b + c) / 2;
            double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return area;
        }

        double area = CalculateTriangleArea(A, B, C);

        Console.WriteLine($"Area of the triangle with sides a={A}, b={B}, and c={C} is is {area}");



    }
}
```

## Question 4 C#
Write a program in C# Sharp to separate odd and even integers in separate arrays.
Go to the editor
Test Data 
:Input the number of elements to be stored in the array :5
Input 5 elements in the array :
element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56
element - 4 : 32
Expected Output:The Even elements are:42 56 32
The Odd elements a:
25 47

```python
using System;

class Program
{
    static void Main(string[] args)
    {
        Console.Write("Enter the number of elements to be stored in the array: ");
        int total_num_elements = int.Parse(Console.ReadLine());
        int[] num_array = new int[total_num_elements];
        int[] even_num_array = new int[total_num_elements];
        int[] odd_num_array = new int[total_num_elements];


        Console.WriteLine($"Input {total_num_elements} elements into the array:");

        for (int i = 0; i < total_num_elements; i++)
        {
            Console.Write($"Element {i + 1}: ");
            num_array[i] = int.Parse(Console.ReadLine());
        }

        int even_count = 0;
        int odd_count = 0;

        foreach (int num in num_array)
        {
            if (num % 2 == 0)
            {
                even_num_array[even_count] = num;
                even_count++;
            }
            else
            {
                odd_num_array[odd_count] = num;
                odd_count++;
            }
        }

        Console.WriteLine("Even elements:");

        for (int i = 0; i < even_count; i++)
        {
            Console.Write(even_num_array[i] + " ");
        }

        Console.WriteLine("\nOdd elements:");

        for (int i = 0; i < odd_count; i++)
        {
            Console.Write(odd_num_array[i] + " ");
        }
    }
}
```

```python
using System;

class Program
{
    static void Main(string[] args)
    {
        bool IsPointInsideRectangle(double x, double y, double x1, double y1, double x2, double y2)
        {
            double lowerLeftCornerX = x - x1;
            double lowerLeftCornerY = y - y1;
            double upperRightCornerX = x2 - x;
            double upperRightCornerY = y2 - y;
            bool check = false;

            if ((lowerLeftCornerX > 0 && lowerLeftCornerY > 0) && (upperRightCornerX > 0 && upperRightCornerY > 0))
            {
                check = true;
            }
            else
            {
                check = false;
            }

            return check;
        }

        bool isInsideRectangle1 = IsPointInsideRectangle(1, 1, 0.3, 0.5, 1.1, 0.7);
        bool isInsideRectangle2 = IsPointInsideRectangle(1, 1, 0.5, 0.2, 1.1, 2);

        if (isInsideRectangle1 && isInsideRectangle2)
        {
            Console.WriteLine("(1,1) lies in both of the given rectangles.");
        }
        else
        {
            Console.WriteLine("(1,1) doesn't lie in both of the given rectangles.");
        }
    }
}
```
